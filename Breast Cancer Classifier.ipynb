```python

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from matplotlib import pyplot as plt
# load data
breast_cancer_data = load_breast_cancer()
# print(breast_cancer_data.data[0])
# print(breast_cancer_data.feature_names)
# print(breast_cancer_data.target)
# print(breast_cancer_data.target_names)
# seperate data into train and test data
training_data, validation_data,training_labels,validation_labels = train_test_split(breast_cancer_data.data,breast_cancer_data.target,test_size =0.2, random_state = 190) 
# confirm effectivness of split function
print(len(training_data )== len(training_labels))
# classifier object
classifier = KNeighborsClassifier(n_neighbors = 3)
# train classifier
classifier.fit(training_data,training_labels)

# check score 
print(classifier.score(validation_data,validation_labels))
# checking for a better k
k_list = range (1,101)
accuracies = []
for i in range (1,101):
  classifier = KNeighborsClassifier(n_neighbors = i)
  classifier.fit(training_data,training_labels)
  # print(classifier.score(validation_data,validation_labels))
  accuracies.append(classifier.score(validation_data,validation_labels))
plt.plot(k_list,accuracies)
plt.xlabel("k")
plt.ylabel("Validation Accuracy")
plt.title("Breast Cancer Classifier Accuracy")
plt.show()
```
